```python
# Build new dictionary to be used to categorize data column after
location_dict = attritions["Location"].value_counts()
print(location_dict)

location_dict_new = {
    'Chennai':       7,
    'Noida':         6,
    'Bangalore':     5,
    'Hyderabad':     4,
    'Pune':          3,
    'Madurai':       2,
    'Lucknow':       1,
    'other place':         0,
}

print(location_dict_new)
```

```python
def location(x):
    if str(x) in location_dict_new.keys():
        return location_dict_new[str(x)]
    else:
        return location_dict_new['other place']
    
data_l = attritions["Location"].apply(location)
attritions['New Location'] = data_l
attritions.head()
```

```python
# Marital Status
# Marital Status
def Mar(x):
    if str(x) in maritial_dict.keys() and maritial_dict[str(x)] > 100:
        return str(x)
    else:
        return 'other status'
    
data_l = attritions["Marital Status"].apply(Mar)
attritions['New Marital'] = data_l
attritions.head()
```

```python
#Promoted/Non Promoted
#Promoted/Not Promoted
def Promoted(x):
    if x == 'Promoted':
        return int(1)
    else:
        return int(0)
    
data_l = attritions["Promoted/Non Promoted"].apply(Promoted)
attritions['New Promotion'] = data_l
attritions.head()
```

```python
# Employee Group
Emp_dict_new = {
    'B1': 4,
    'B2': 3,
    'B3': 2,
    'other group': 1,
}


def emp(x):
    if str(x) in Emp_dict_new.keys():
        return str(x)
    else:
        return 'other group'
    
data_l = attritions["Emp. Group"].apply(emp)
attritions['New EMP'] = data_l

emp = pd.get_dummies(attritions["New EMP"])
attritions.head()
```

```python
#Job Role Match
def Job(x):
    if x == 'Yes':
        return int(1)
    else:
        return int(0)
    
data_l = attritions["Job Role Match"].apply(Job)
attritions['New Job Role Match'] = data_l
attritions.head()
```

```python
#Gender
# Gender
def Gen(x):
    if x in gender_dict.keys():
        return str(x)
    else:
        return 'other'
    
data_l = attritions["Gender "].apply(Gen)
attritions['New Gender'] = data_l
attritions.head()
```

```python
dataset=pd.read_csv("processed table.csv")
dataset=pd.DataFrame(dataset)
y = dataset['Stay/Left']
X=dataset.drop("Stay/Left",axis=1)
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=4
)
```

```python
X_train
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import GaussianNB
```

```python
lr=LogisticRegression(C = 0.1, random_state=42 , solver='liblinear')
dt=DecisionTreeClassifier()
rm=RandomForestClassifier()
gnb=GaussianNB()
```

```python
for a,b in zip([lr,dt,rm,gnb],["Logistic Regression","Decision Tree","Random Forest","Naive Bayes"]):
    a.fit(X_train,y_train)
    prediction=a.predict(X_train)
    y_pred=a.predict(X_test)
    score1=accuracy_score(y_train,prediction)
    score=accuracy_score(y_test,y_pred)
    msg1="[%s] training data accuracy is : %f" % (b,score1)
    msg2="[%s] test data accuracy is : %f" % (b,score)
    print(msg1)
    print(msg2)
```

```python
model = lr.fit(X_train,y_train)

import pickle
# save the model to disk
filename = 'finalized_model.pickle'
pickle.dump(model, open(filename, 'wb'))
```
